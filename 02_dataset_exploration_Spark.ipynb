```python
import spark

path="/Volumes/workspace/default/dataset/mhd2.csv"


df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(path)
display(df)
```

```python
from pyspark.sql.functions import count

dfGroupByGender = df.groupBy("Gender").agg(count(df["Gender"]))

display(dfGroupByGender)
```

```python
from pyspark.sql.functions import count

dfGroupByGender = df.groupBy("Gender","Mood_label").agg(count(df["Gender"]))

display(dfGroupByGender.orderBy("Mood_label","Gender"))

```

```python
dfMaleMood = df[df["Gender"]=="Male"].groupBy("Mood_label").agg(count(df["*"]))
display(dfMaleMood.orderBy("Mood_label"))
```

```python
dfPazientiGravi = df[(df["Anxiety_Score (0-100)"]>80) | (df["Depression_Score (0-100)"]>80) | (df["Stress_Score (0-100)"]>80)]
display(dfPazientiGravi)
```

```python
from pyspark.sql.functions import count

dfGroupByMentalCondition = df.groupBy("Mental_Health_Condition").agg(count(df["*"]))

display(dfGroupByMentalCondition)
```
